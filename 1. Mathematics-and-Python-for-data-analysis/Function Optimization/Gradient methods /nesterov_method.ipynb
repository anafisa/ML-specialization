# Метод Нестерова 

## Реализация метода Нестерова

```python
def nest_method(f, x0, y0, eta, gamma, e, v_x=0, v_y=0, i=0): 
    
    g = grad(f, x0, y0)
    x1 = x0 - eta * g[0]
    y1 = y0 - eta * g[1]
    
    while abs(f(x1, y1) - f(x0, y0)) > e:
        v_x = gamma * v_x + eta * grad(f, x1 - gamma * v_x, y1 - gamma * v_y)[0] 
        v_y = gamma * v_y + eta * grad(f, x1 - gamma * v_x, y1 - gamma * v_y)[1]
        x0, y0 = x1, y1
        x1 -= v_x
        y1 -= v_y
        i += 1
        
    return (x1, y1), i
```

```python
def f(x, y):
    return (1.5 - x + x*y)**2 + (2.25 - x + x*(y**2))**2 + (2.625 - x + x*(y**3))**2

def dx(f, x, y):
    return 2 * (-1 + y) * (1.5 - x + x*y) + 2*(-1 + y**2)*(2.25 - x + x*y**2) + 2*(-1 + y**3)*(2.625 - x + x*(y**3))

def dy(f, x, y):
    return 2*x*(1.5 - x + x*y) + 4 *x *y* (2.25 - x + x* y**2) + 6* x* y**2 *(2.625 - x + x * y**3)

def grad(f, x0, y0):
    d_x = dx(f, x0, y0)
    d_y = dy(f, x0, y0)
    return d_x, d_y
```

```python
nest_method(f, 0.7, 0.4, 0.01, 0.9, 0.0001)
```
