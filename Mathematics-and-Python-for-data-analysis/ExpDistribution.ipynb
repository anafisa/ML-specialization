# Экспоненциальное распределение

* [подробнее о распределении](https://ru.wikipedia.org/wiki/Экспоненциальное_распределение)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

1) Сгенерировать выборку из экспоненциального распределения размером 1000, построить гистограмму выборки и нарисовать поверх неё теоретическую плотность распределения случайной величины 

```python
# задаем параметр распределения
lamb = 1 
exp_rv = sts.expon(lamb)

# генерируем выборку
sample = exp_rv.rvs(size=1000) 
```

```python
# строим гистограмму по выборке
plt.hist(sample, density=True, label='Real distribution') 
x = np.linspace(1,10,100) 

# строим теоритическую плотность
pdf = exp_rv.pdf(x)
plt.plot(x, pdf, label='Theoretical pdf', alpha=1)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

2) Оценить распределение выборочного среднего случайной величины при разных объёмах выборок. Для этого при трёх и более значениях n (например, 5, 10, 50) необходимо сгенерировать 1000 выборок объёма n и построить гистограммы распределений их выборочных средних. 

3)  Используя информацию о среднем и дисперсии исходного распределения расчитать параметры нормальных распределений.

```python
# рассчет выборочных средних n = 5
sample = []
[sample.append(exp_rv.rvs(size=5).mean()) for i in range(1000)]

# строим гистограмму по выборочным средним
plt.hist(sample, density=True, label='Mean distribution') 

# строим теоритическую плотность
means = np.array(sample)
mean = means.mean()
sd = means.std()
norm_rv = sts.norm(mean, sd)
x = np.linspace(min(means), max(means), 1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='Theoretical pdf', alpha=1)
plt.text(3.1, 0.5, "$\sigma^2 = {}$\n$\mu={}$".format(round(sd**2, 2), round(mean, 2)),
             size = 13,
             bbox=dict(facecolor='white', alpha=0.5))
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
